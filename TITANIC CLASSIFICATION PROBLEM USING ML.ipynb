## Importing Data and Testing Files

```python
t_train = pd.read_csv('F:/DATA SCIENCE/PROJECT 4 TITANIC SURVIVALS/train.csv')
t_test = pd.read_csv('F:/DATA SCIENCE/PROJECT 4 TITANIC SURVIVALS/test.csv')

#The above code loads Data File(train) and Test File (test) into the Jupyter.
```

```python
print(t_train.columns)
print(t_test.columns)
print(t_train.shape)
print(t_test.shape)

#Displays the details of columns and shape 
```

```python
t_train['CabinAvailability'] = t_train["Cabin"].apply(lambda x: 0 if type(x) == float else 1)
# Filling empty cell of Embarked with its mode value

t_train.head()
```

# Finding the relationship of various factors that affected surviving by calculating the mean of survival based on categories in each column (higher the number, more chances of surviving

```python
t_train[['Cabin', 'Survived']].groupby(['Cabin'], as_index=False).mean()

# Survivors w.r.t Pclass
```

```python
t_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean()

# Survivors w.r.t Pclass
```

```python
t_train[["Sex", "Survived"]].groupby(['Sex'], as_index=False).mean()

# Survivors w.r.t Sex
```

```python
replacement = {
    'female': 0,
    'male': 1,
    
}

t_train['Sex'] = t_train['Sex'].apply(lambda x: replacement.get(x))
t_train.head()
```

# From the data available, Family size might have played a role in survival. Familiy size= Passenger+SibSp+ParCh. Create a new column: FamilySize

```python
t_train["FamilySize"] = 1 + t_train["SibSp"] + t_train["Parch"]
```

```python
t_train[["FamilySize","SibSp","Parch"]].head(10)
```

```python
t_train[["FamilySize", "Survived"]].groupby(['FamilySize'], as_index=False).mean()

# Survivors w.r.t Sex
```

```python
replacement = {
    4: 0,
    3: 1,
    2: 1,
    7: 2,
    1: 2,
    5: 3,
    6: 4,
    8: 5,
    11:5
    
    
}

t_train['FamilySize'] = t_train['FamilySize'].apply(lambda x: replacement.get(x))
t_train.head()
```

```python
t_train["IsAlone"]=0
t_train.loc[t_train['FamilySize'] == 1, 'IsAlone'] = 1
t_train.head()

# Creating a new column to set if the passenger is alone or not
```

```python
for col in ["Cabin"]:
    t_train[col].fillna('NoValue',inplace=True)
# Filling empty cell of Embarked with its mode value

t_train.isnull().sum().sort_values(ascending=False)
```

```python
for col in ["Embarked"]:
    t_train[col].fillna(t_train[col].value_counts().idxmax(),inplace=True)
# Filling empty cell of Embarked with its mode value

t_train.isnull().sum().sort_values(ascending=False)
```

```python
t_train[["Embarked", "Survived"]].groupby(['Embarked'], as_index=False).mean()

# Survivors w.r.t Embarked
```

```python
replacement = {
    "C": 0,
    "Q": 1,
    "S": 2,
    
}

t_train['Embarked'] = t_train['Embarked'].apply(lambda x: replacement.get(x))
t_train.head()
```

# Empty cells in Ages will be filled with random numbers between
# (avg age + std dev) and (avg age - std dev)

```python
age_avg = t_train['Age'].mean()
age_std = t_train['Age'].std()
age_null_count = t_train['Age'].isnull().sum()

# Calculated mean and deviation to fill Nans, Empty cells number is taken to generate that many random numbers.
```

```python
age_avg, age_std, age_null_count
```

```python
age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
# A random numbers for age is generated
```

```python
age_null_random_list
```

```python
t_train['Age'][np.isnan(t_train['Age'])] = age_null_random_list
t_train['Age'] = t_train['Age'].astype(int)

# The random age generated is filled in data set wherever ther is NaNs
```

```python
t_train.isnull().sum().sort_values(ascending=False)
```

```python
t_train['CategoricalAge'] = pd.qcut(t_train['Age'], 5)
print (t_train[['CategoricalAge', 'Survived']].groupby(['CategoricalAge'], as_index=False).mean())
```

```python
t_train.loc[(t_train['Age'] > -1) & (t_train['Age'] <= 19.0), 'Age'] = 0
t_train.loc[(t_train['Age'] > 19.0) & (t_train['Age'] <= 25.0), 'Age'] = 1
t_train.loc[(t_train['Age'] > 25.0) & (t_train['Age'] <= 32.0), 'Age'] = 2
t_train.loc[(t_train['Age'] > 32.0) & (t_train['Age'] <= 40.0), 'Age'] = 3
t_train.loc[ t_train['Age'] > 40.0, 'Age'] = 4
t_train['Age'] = t_train['Age'].astype(int)
```

```python
t_train[["Age", "Survived"]].groupby(['Age'], as_index=False).mean()

# Survivors w.r.t Embarked
```

```python
replacement = {
    0: 0,
    2: 1,
    3: 2,
    4: 2,
    1: 3
    
}

t_train['Age'] = t_train['Age'].apply(lambda x: replacement.get(x))
t_train.head()
```

# From the names, Titles might have played a role in survival, Create a new column Title

```python
t_train["Title"]=0
t_train.head()
```

```python
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
# If the title exists, extract and return it.

    if title_search:
        return title_search.group(1)
    return ""
```

```python
titles = t_train["Name"].apply(get_title)

# Get all the titles
```

```python
t_train["Title"] = titles

# Creating a new Column Titles in the  DF
```

```python
t_train['Title'] = t_train['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
```

```python
print (t_train[['Title', 'Survived']].groupby(['Title'], as_index=False).mean())
```

```python
replacement = {
    "Mrs": 0,
    "Miss": 1,
    "Master": 2,
    "Rare": 3,
    "Mr": 3,
    "Mlle": 4,
    "Mme": 4,
    "Ms": 4
    
}

t_train['Title'] = t_train['Title'].apply(lambda x: replacement.get(x))
t_train.head()
```

```python
t_train.isnull().sum().sort_values(ascending=False)
```

# Since most of the Cabin Values are missing, Fare must be related to type of cabin as well as place of cabin to survive. Categorise fare

```python
t_train['CategoricalFare'] = pd.qcut(t_train['Fare'], 4)

print (t_train[['CategoricalFare', 'Survived']].groupby(['CategoricalFare'], as_index=False).mean())
```

```python
t_train.loc[(t_train['Fare'] > -1) & (t_train['Fare'] <= 7.91), 'Fare'] = 3
t_train.loc[(t_train['Fare'] > 7.91) & (t_train['Fare'] <= 14.454), 'Fare'] = 2
t_train.loc[(t_train['Fare'] > 14.454) & (t_train['Fare'] <= 31.0), 'Fare'] = 1
t_train.loc[ t_train['Fare'] > 31.0, 'Fare'] = 0
t_train['Fare'] = t_train['Fare'].astype(int)

# Categorising Fare column w.r.t Categorical Fare developed
```

```python
t_train.head()
```

```python
del t_train["Name"]
del t_train["Ticket"]
del t_train["CategoricalFare"]
del t_train["CategoricalAge"]
del t_train["Parch"]
del t_train["SibSp"]
del t_train["Cabin"]
del t_train["PassengerId"]

# Deleting Name column as the tiltes has been taken, and also ticket and categorical as the fares has been categorised.
```
